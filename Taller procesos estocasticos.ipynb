  1. Dada una variable aleatoria $X =  Gamma(k,\theta)$ con parametros de forma $k > 0$, prametros de escala  $\theta > 0$      y con FDP dada por:
  
      $$F_X(x) =  \frac {x^{k-1}e^{\frac{-x}{\theta}}}{\theta^{k}\Gamma(k)}$$
     
     donde $\Gamma(k)$ es la función gamma dada por
     
     $$\Gamma(k) = \int_0^\infty \ u^{k-1}e^{-u}du$$
     
     Demuestre matematicamente que:
     
     - $E[X] = k\theta$
     - $Var[X] =  k\theta^2$
     
     ##Demostración
     ###$E[X]=k\theta$
     $$E[X]=\int_{-\infty}^{\infty} \ xf(x) = \int_0^\infty \ x \frac {x^{k-1}e^{\frac{-x}{\theta}}}{\theta^k\Gamma(k)}dx$$
     $$\frac{1}{\theta^k\Gamma(k)}\int_0^\infty \ x^{k}e^{\frac{-x}{\theta}}dx$$
     remplazamos a $u=\frac{x}{\theta}$, $x=u\theta$ y $dx=\theta du$ entonces:
     $$\frac{1}{\theta^k\Gamma(k)}\int_0^\infty \ (u\theta)^{k}e^{-u}\theta du$$
     $$\frac{1}{\theta^k\Gamma(k)}\int_0^\infty \ u^{k}\theta^{k}e^{-u}\theta du$$
     $$\frac{\theta^k\theta}{\theta^k\Gamma(k)}\int_0^\infty \ u^{k}e^{-u}du$$
     la integral $\int_0^\infty \ u^{k}e^{-u}du = \Gamma(k+1)$ y $\Gamma(k+1) = k\Gamma(k)$ entonces:
     $$\frac{\theta}{\Gamma(k)} \Gamma(k+1) = \frac{\theta k}{\Gamma(k)} \Gamma(k) = \theta k$$
     
     ###$Var[X]=k\theta^2$
     La Varianza esta dada por la siguiente expresión $Var(X)=E[X^2]-(E[X])^2$ entonces se haya $E[X^2]$
     $$E[X^2]=\int_{-\infty}^{\infty} \ x^2f(x) = \int_0^\infty \ x^2 \frac {x^{k-1}e^{\frac{-x}{\theta}}}{\theta^k\Gamma(k)}dx$$
     $$\frac{1}{\theta^k\Gamma(k)}\int_0^\infty \ x^{k+1}e^{\frac{-x}{\theta}}dx$$
     remplazamos a $u=\frac{x}{\theta}$, $x=u\theta$ y $dx=\theta du$ entonces:
     $$\frac{1}{\theta^k\Gamma(k)}\int_0^\infty \ (u\theta)^{k+1}e^{-u}\theta du$$
     $$\frac{\theta^{k+1}\theta}{\theta^k\Gamma(k)}\int_0^\infty \ u^{k+1}e^{-u}du$$
     $$\frac{\theta^{k+2}}{\theta^k\Gamma(k)}\int_0^\infty \ u^{k+1}e^{-u}du$$
     $$\frac{\theta^{k}\theta^2}{\theta^k\Gamma(k)}\int_0^\infty \ u^{k+1}e^{-u}du$$
     $$\frac{\theta^2}{\Gamma(k)}\int_0^\infty \ u^{k+1}e^{-u}du$$
     $$\frac{\theta^2}{\Gamma(k)}\Gamma(k+2)=\frac{\theta^2 (k+1)k\Gamma(k)}{\Gamma(k)}=\theta^2(k+1)k$$
     entonces $E[X^2]=\theta^2(k+1)k$ remplazamos en la ecuacion inicial para hayar la varianza:
     $$Var(X)=E[X^2]-(E[X])^2=\theta^2(k+1)k-(\theta k)^2=\theta^2(k^2+k)-\theta^2 k^2=\theta^2k^2+\theta^2k-\theta^2 k^2=\theta^2k$$
     
   
   2. Dada una variable aleatoria $X \sim Exp(\theta)$ con parametros  de escala $\theta > 0$ y con FDP dada por:
   
   $$f_X(x) =  \frac {1}{\theta} e^{\frac{-x}{\theta}}$$
   
     Demuestre matematicamente que:
     - $E[X] = \theta$
     - $Var[X] = \theta^2$
     
     ##Demostración
     ###$E[X] = \theta$
     $$E[X]=\int_0^\infty \ xf(x) = \int_0^\infty \ x\frac{1}{\theta}e^{-\frac{x}{\theta}}dx$$
     $$E[X]= \frac{1}{\theta}\int_0^\infty \ xe^{-\frac{x}{\theta}}dx$$
     remplazamos a $u=\frac{x}{\theta}$, $x=u\theta$ y $dx=\theta du$ entonces:
     $$E[X]=\frac{1}{\theta}\int_0^\infty \ (u\theta)e^{-u}\theta du$$
     $$E[X]= \frac{1}{\theta}\int_0^\infty \ u\theta^2e^{-u} du$$
     $$E[X]=\frac{\theta^2}{\theta}\int_0^\infty \ ue^{-u} du$$
     $$E[X]= \theta\int_0^\infty \ ue^{-u} du$$
     $$E[X]= \theta[ ue^{-u} - \int_0^\infty \ e^{-u}du]$$
     $$E[X]= \theta[ ue^{-u} + e^{-u}]$$
     $$E[X]= \theta[ (0e^{-0} + e^{-0})+ \lim\limits_{u\rightarrow\infty} ue^{-u} + e^{-u}]$$
     $$E[X]= \theta$$
     
     ###$Var[X] = \theta^2$
     La Varianza esta dada por la siguiente expresión $Var(X)=E[X^2]-(E[X])^2$ entonces se haya $E[X^2]$
     $$E[X^2]=\int_{-\infty}^{\infty} \ x^2f(x) = \int_0^\infty \ x^2 \frac{1}{\theta}e^{-\frac{x}{\theta}}dx$$
     $$E[X^2]=\frac{1}{\theta}\int_0^\infty \ x^2e^{-\frac{x}{\theta}}dx$$
     remplazamos a $u=\frac{x}{\theta}$, $x=u\theta$ y $dx=\theta du$ entonces:
     $$E[X^2]=\frac{1}{\theta}\int_0^\infty \ (u\theta)^2e^{-u}\theta du$$
     $$E[X^2]=\frac{1}{\theta}\int_0^\infty \ u^2\theta^3e^{-u} du$$
     $$E[X^2]=\frac{\theta^3}{\theta}\int_0^\infty \ u^2e^{-u} du$$
     $$E[X^2]=\theta^2\int_0^\infty \ u^2e^{-u} du$$
     $$E[X^2]=\theta^2[e^-u(-u^2-2u-2)]=2\theta^2$$
     $$Var(X)=E[X^2]-(E[X])^2=2\theta^2-\theta^2=\theta^2$$

  4. Dada $X_1, X_2,..., X_n$ variables aleatorias muruamente idependientes (discretas o continuas), cada una con media $\mu$ y varianza $\sigma^2$ finitas. Entonces  si $S_n = X_1+X_2+...+X_n$ con $n=1,2,3,...$, demuestre que
  $$\lim\limits_{n\rightarrow \infty} P(|\frac {S_n}{n}-\mu|>= \epsilon)=0$$
  ##Demostración:
  $$E(\frac{S_n}{n})=\frac {1}{n} \sum\limits_{i=1}^{n} E(X_i) = \frac {1}{n}n, \mu = \mu$$
  $$Var(\frac{S_n}{n})=\frac{1}{n^2}Var(\sum\limits_{i=1}^{n} X_i)=\frac{1}{n^2}\sum\limits_{i=1}^{n}Var(X_i)=\frac {n\sigma^2}{n^2}=\frac {\sigma^2}{n}$$
  por lo tanto, $P(|X-E(X)|\geq \epsilon\leq\frac{1}{\epsilon^2}Var(X))$
  $$\lim\limits_{n\rightarrow\infty} P(|\frac {1}{n}\sum\limits_{i=1}^{n} X_i-\mu|\geq\epsilon)\leq\frac {\sigma^2}{n\epsilon^2}$$
  Entonces converge $n\rightarrow\infty$, $\frac {\sigma^2}{n\epsilon^2}\rightarrow0$ 
  5. Dadas $\{Y_n : n \geq 1\}$ una secuencia de variables aleatorias  y ausmiendo que $\lim\limits_{n\rightarrow \infty} E[|Y_n|]=0$. Demuestre que la probabilidad  de $\{Y_n : n \geq 1\}$ converge a 0.
  $$P\{Y_n: n\geq1\}$$
  ##Demostración:
  Usando la desigualdad de markov que muestra $P(|X|\geq a) \leq \frac{E(X)}{a}$
  entonces:
  
  $$P(Y_n:n \geq 1)\leq\frac {E(Y_n)}{n}$$ 
  
  entonces conforma $n\rightarrow\infty$, $\frac {E(Y_n)}{n}\rightarrow0$.

```python
import numpy as np
import scipy.stats as ss
import pylab as plt
import matplotlib

plt.figure(figsize=(17,16))
alpha=5
loc=100.5
beta=22
x=ss.gamma.rvs(alpha,loc=loc,scale=beta,size=10000)
expectedValueT=beta*alpha
mean=x.mean()#promedio o valor esperado
bins=100
siz,bins,patches=plt.hist(x,bins,histtype='stepfilled',label='sample gamma data')
plt.setp(patches,'facecolor','g','alpha',0.75)
plt.grid()
plt.plot([mean,mean],[0,max(siz)],color='r',label="Sample Mean:  "+str(mean))
plt.plot([expectedValueT,expectedValueT],[0,max(siz)],color="b",label="Expected Value"+str(expectedValueT))
plt.legend()
plt.show()
```
